```python
import pandas as pd
```

```python
data = pd.DataFrame()
for num_annotators in [2,3]:
    for glove in ["twit", "wiki", "no"]:
        for glove_dim in [100, 200]:
            for num_layers in [1,2]:
                for num_nodes in [128,256]:
                    for rnn in ["lstm", "gru"]:
                        for bidirectional in [True, False]:
                            bid = ""
                            if bidirectional:
                                    bid = "_bid"
                            for pyramid in [True, False]:
                                pyr = ""
                                if pyramid:
                                    pyr = "_pyr"
                                pathname = "ann"+str(num_annotators)+"_"+glove+"_gdim"+str(glove_dim)+"_"+rnn+bid+"_nlayers"+str(num_layers)+"_nnodes"+str(num_nodes)+pyr
                                data_loc = pd.read_csv("./experiments_01/"+pathname+"/evaluation.csv")
                                data_loc["exp"] = pathname
                                data_loc["info"] = ["precision", "recall", "f1", "none", "time"]
                                data_loc["ann"] = [num_annotators]*len(data_loc)
                                data_loc["glove"] = [glove]*len(data_loc)
                                data_loc["gdim"] = [glove_dim]*len(data_loc)
                                data_loc["rnn"] = [rnn]*len(data_loc)
                                data_loc["bid"] = [bidirectional]*len(data_loc)
                                data_loc["nl"] = [num_layers]*len(data_loc)
                                data_loc["nn"] = [num_nodes]*len(data_loc)
                                data_loc["pyr"] = [pyramid]*len(data_loc)
                                data = data.append(data_loc)
data = data.rename(columns={"0": "ns", "1": "sens"})
data
```

```python
main_cols = ["accuracy", "ann", "glove", "gdim", "rnn", "bid", "nl", "nn", "pyr"]
data[data["ann"] == 2][main_cols].drop_duplicates().sort_values(by="accuracy", ascending=False).reset_index(drop=True)[:10]
```

```python
def print_top(data, top_values):
    main_cols = ["ns", "sens", "accuracy", "ann", "glove", "gdim", "rnn", "bid", "nl", "nn", "pyr"]
    data_loc = data[(data["info"] == "f1")][main_cols].drop_duplicates().sort_values(by="accuracy", ascending=False).reset_index(drop=True)
    print("##### Top", top_values, "models for accuracy")
    print(data_loc[:10])
    for cls in ["ns", "sens"]:
        print("\n##### Top", top_values, "models for f1 on", cls)
        print(data_loc.sort_values(by=cls, ascending=False)[:10])
    return

for ann in [2,3]:
    print("\n##############", ann, "annotators")
    print_top(data[(data["ann"] == ann)], 10)
```
