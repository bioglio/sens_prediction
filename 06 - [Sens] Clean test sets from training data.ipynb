## Create new test sets

```python
df_tr2 = pd.read_csv(DATA_PATH+"annotation_results__ann2_training.csv")
df_te2 = pd.read_csv(DATA_PATH+"annotation_results__ann2_test.csv")
df_tr3 = pd.read_csv(DATA_PATH+"annotation_results__ann3_training.csv")
df_te3 = pd.read_csv(DATA_PATH+"annotation_results__ann3_test.csv")
```

```python
tmp = df_te3.merge(df_tr2, left_on="text", right_on="text", how="left", suffixes=('', '_r'))
tmp = tmp[tmp['class_r'].isna()].drop(columns=["class_r"])
tmp.to_csv(DATA_PATH+"annotation_results__ann3_test-notrains2.csv")
tmp
```
