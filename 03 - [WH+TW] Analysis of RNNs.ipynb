```python
import pandas as pd
```

```python
# directories
main_path = "./"
dataset_filename = main_path+"data/annotation_results__ann"
dataset_wt_filename = main_path+"data/sample_ann2_"
glove_path = main_path+"glove/"
results_path = main_path+"experiments_02/"
results_path_bow = main_path+"experiments_03/"
models_path = results_path+"models/"
stats_path = results_path+"results/"
Y_feat_names = ["ns", "sens"]

# variables of NNs
ann_list = [2, 2, 3, 3]
glove_dim_list = [100, 100, 200, 200]
rnn_list = ["gru", "lstm", "gru", "lstm"]
bidirectional_list = [True, True, True, True]
num_layers_list = [2, 2, 1, 2]
num_nodes_list = [128, 256, 128, 128]
pyramid_list = [False, True, False, False]
glove = "twit"
```

```python
df_stats = pd.DataFrame()
for ann, glove_dim, rnn, bidirectional, num_layers, num_nodes, pyramid in zip(ann_list, glove_dim_list, rnn_list, bidirectional_list, num_layers_list, num_nodes_list, pyramid_list):
    # other
    bid = ""
    if bidirectional:
        bid = "_bid"
    pyr = ""
    div_nodes = 0
    if pyramid:
        pyr = "_pyr"
        div_nodes = 2
    # load data
    for index in range(10):
        model_name = "sample"+(("0"+str(index+1))[-2:])+"_ann"+str(ann)+"_"+glove+"_gdim"+str(glove_dim)+"_"+rnn+bid+"_nlayers"+str(num_layers)+"_nnodes"+str(num_nodes)+pyr
        # get data
        df_stats_loc = pd.read_csv(stats_path+model_name+".csv")
        df_stats_loc["set"] = "sample_"+str(("0"+str(index+1))[-2:])
        df_stats_loc["data"] = "ann"+str(ann)
        #df_stats_loc["mod"] = model_name
        df_stats_loc["mod"] = rnn
        df_stats = df_stats.append(df_stats_loc)
```

## Gain over BoW

```python
met_dict_df_bow = pd.read_csv(results_path_bow+"bow_samples.csv")
met_dict_df_bow = met_dict_df_bow[(met_dict_df_bow["mod"] == "lSVC") & (met_dict_df_bow["set"] == "test")]
#met_dict_df_bow
df_res = df_stats[["f1-macro", "set", "data", "mod"]].merge(met_dict_df_bow[["f1-macro", "set", "data", "mod"]], left_on="set", right_on="data", suffixes=('', '_bow'))
#df_res
df_res["gain"] = (df_res["f1-macro"]/df_res["f1-macro_bow"] - 1.0)*100
#df_res
df_res[["data", "mod", "gain"]].groupby(["data", "mod"]).agg({"gain":['mean','std']}).reset_index()
```
